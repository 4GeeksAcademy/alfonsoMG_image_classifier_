```python
import zipfile
import os
import shutil
import matplotlib.pyplot as plt
from PIL import Image
import random
```

```python
# Ruta al archivo zip
zip_file_path = '../data/raw/train.zip'

# Ruta de destino para la extracción
extraction_path = '../data/processed/train_extracted'


# Ruta al archivo zip
zip_file_path_test = '../data/raw/test1.zip'

# Ruta de destino para la extracción
extraction_path_test = '../data/interim/test'


# Crea la carpeta de destino si no existe
os.makedirs(extraction_path, exist_ok=True)

# Descomprime el archivo zip
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extraction_path)

# Crea la carpeta de destino si no existe
os.makedirs(extraction_path_test, exist_ok=True)

# Descomprime el archivo zip
with zipfile.ZipFile(zip_file_path_test, 'r') as zip_ref:
    file_list = zip_ref.namelist()
    files_to_extract = random.sample(file_list, int(0.1 * len(file_list)))
    zip_ref.extractall(extraction_path_test, members=files_to_extract)
```

```python
total_animals = "../data/processed/train_extracted/train"
animals = os.listdir(total_animals)
print(len(animals))
```

```python
# Ruta de la carpeta que contiene las imágenes
carpeta_imagenes = '../data/processed/train_extracted/train'

# Obtener la lista de archivos en la carpeta
archivos = os.listdir(carpeta_imagenes)

# Filtrar archivos que contienen la palabra "cat"
archivos_cat = [archivo for archivo in archivos if 'cat' in archivo]

# Calcular el 5% de los archivos
porcentaje_a_copiar = int(0.05 * len(animals))

# Seleccionar aleatoriamente el 5% de los archivos
archivos_cat_a_copiar = random.sample(archivos_cat, porcentaje_a_copiar)

nueva_carpeta = '../data/interim/cat'

# Crear la nueva carpeta si no existe
if not os.path.exists(nueva_carpeta):
    os.makedirs(nueva_carpeta)

# Copiar los archivos a la nueva carpeta
for archivo_cat in archivos_cat_a_copiar:
    ruta_origen = os.path.join(carpeta_imagenes, archivo_cat)
    ruta_destino = os.path.join(nueva_carpeta, archivo_cat)
    shutil.copy2(ruta_origen, ruta_destino)

# Contador para limitar a las primeras 9 imágenes con nombres que contienen "cat"
contador_cat = 0

# Iterar sobre cada archivo en la carpeta
for archivo in archivos:
    # Comprobar si el archivo es una imagen y contiene "cat" en el nombre
    if contador_cat < 9 and archivo.lower().endswith(('.jpg')) and "cat" in archivo.lower():
        # Incrementar el contador
        contador_cat += 1

        # Crear la ruta completa del archivo
        ruta_completa = os.path.join(carpeta_imagenes, archivo)

        # Abrir la imagen con Pillow
        imagen = Image.open(ruta_completa)

        # Mostrar la imagen
        plt.subplot(3, 3, contador_cat)
        plt.imshow(imagen)
        plt.title(f"Imagen {contador_cat}")
        plt.axis("off")

# Ajustar el diseño y mostrar las imágenes
plt.tight_layout()
plt.show()
```
